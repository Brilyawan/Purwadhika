### QUESTION #1

1.1. Create a dataframe

```python
data = [
    {'Name': 'Sara', 'Age': 24, 'Gender': 'Female', 'Occupation': 'Data Analyst', 'Domicilie': 'Jakarta'},
    {'Name': 'Faris', 'Age': 36, 'Gender': 'Male', 'Occupation': 'Data Scientist', 'Domicilie': 'Bandung'},
    {'Name': 'Ahmad', 'Age': 22, 'Gender': 'Male', 'Occupation': 'BI Analyst', 'Domicilie': 'Yogyakarta'},
    {'Name': 'Fara', 'Age': 27, 'Gender': 'Female', 'Occupation': 'Data Scientist', 'Domicilie': 'Jakarta'},
    {'Name': 'Nisa', 'Age': 25, 'Gender': 'Female', 'Occupation': 'Data Engineer', 'Domicilie': 'Tangerang'},
    {'Name': 'Catur', 'Age': 30, 'Gender': 'Male', 'Occupation': 'Data Engineer', 'Domicilie': 'Bali'},
    {'Name': 'Jessica', 'Age': 28, 'Gender': 'Female', 'Occupation': 'Data Analyst', 'Domicilie': 'Bekasi'},
]

df = pd.DataFrame(data)
df
```

1.2. Change the order of the columns to be as follows: Name, Gender, Age, Domicilie, Occupation. 

```python
df = df[['Name', 'Gender', 'Age', 'Domicilie', 'Occupation']]
df
```

1.3. Add 1 new column named Hobby and fill in the data.

```python
df['Hobby'] = ['Traveling', 'Gaming', 'Dancing', 'Gaming', 'Traveling', 'Cooking', 'Reading']
df
```

1.4. Grab data on Ahmad, Nisa, and Sara.

```python
df[df.Name.isin(['Ahmad', 'Nisa', 'Sara'])]
```

1.5. Show domicile and occupation of Faris and Jessica (including name field).

```python
df.loc[
    df.Name.isin(['Faris', 'Jessica']),
    ['Name', 'Domicilie', 'Occupation']
]
```

1.6. Who is over 30 years old?

```python
df[df.Age > 30]
```

1.7. Who is between 26 and 30 years old and has a hobby of gaming?

```python
df.loc[
    (df.Age.between(26, 30)) & (df.Hobby == 'Gaming')
]
```

1.8. Show the names of people whose job is Data Analyst! Show the name column only in Series form!

```python
df.loc[
    df.Occupation == 'Data Analyst', 'Name'
]
```

1.9. What do people who live in Jakarta do for a living? Show us their names and occupations!

```python
df.loc[
    df.Domicilie == 'Jakarta', ['Name', 'Occupation']
]
```

1.10. What are the hobbies of the women in the dataset? Show only the Name, Hobby, Age columns in order!

```python
df.loc[
    df.Gender == 'Female', ['Name', 'Hobby', 'Age']
]
```

1.11. Who are older than Nisa? Show name and age columns!

```python
df[df.Name == 'Nisa']['Age'].values[0]
```

```python
df.loc[
    df.Age > df[df.Name == 'Nisa']['Age'].values[0], ['Name', 'Age']
]
```

1.12. Based on the data, are there any women whose hobby is traveling?

```python
df.loc[
    (df.Gender == 'Female') & (df.Hobby == 'Traveling') 
]
```
